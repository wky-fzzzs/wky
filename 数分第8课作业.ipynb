## seaborn

```python
# 饼图进展
import matplotlib.pyplot as plt

x = [0.25,0.25,0.25,0.1,0.15]
y = ['A','B','C','D','E']

plt.pie(
    x=x,
    labels=y,
    autopct='%1.1f%%', # 第一个% 声明的是格式化输出 1 只少有一位 .1f %% 
    pctdistance=0.8 # 标签 显示位置 1 在圆圏上 1以上在圆之外
)
plt.show()
```

```python
# 08.seaborn演示/iris.csv
import seaborn as sns

# 文件读取
# 花瓣长度、花瓣宽度、萼片长度、萼片宽度、物种

path = r'./08.seaborn演示/iris.csv'
iris = pd.read_csv(path)
iris
```

```python
plt.figure(figsize=(10,5))
# 使用sns画散点图
# x指定花萼的长度字段 y指定花萼宽度的字段 data 引入二维表格对象 hue 指明鸢尾花的物种进行分类
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris,hue='species')
```

```python
# sns.boxplot(x='species',y='sepal_width',data=iris,hue='species')
sns.boxplot(x='species',y='sepal_length',data=iris,hue='species')
```

## 直方图

```python
plt.figure(figsize=(10,5))
# 生成一维直方图
sns.histplot(x='sepal_length',data=iris,hue='species')
plt.show
```

```python
# 生成一维直方图
plt.figure(figsize=(10,5))
# kde 表示的是核密度估计，可以更平哈的展示数据的趋势
sns.histplot(x='sepal_length',data=iris,hue='species',kde=True)
plt.show
```

```python
# 生成二维直方图
plt.figure(figsize=(10,5))
sns.histplot(x='sepal_length',data=iris,y='species')
plt.show
```

## 成对图

```python
plt.figure(figsize=(10,5))
# 使用sns画成对图
# x
sns.pairplot(data=iris,hue='species',markers=['o','D','s'],palette='Set1')
plt.show()
```

## 热力图

```python
plt.figure(figsize=(10,5))
my_data = iris.iloc[:,0:-1].corr()
sns.heatmap(data=my_data,annot=True,cmap='viridis')

plt.show()
```
